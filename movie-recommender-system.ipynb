```python
import ast
```

```python
def convert(obj):# here it would be string
    L=[]
    for i in ast.literal_eval(obj): # here with the help of literal_eval funciton present in ast string->list
        L.append(i['name'])
    return L
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
movies['cast'][0]
```

```python
def convert3(obj):
    L=[]
    count=0
    for i in ast.literal_eval(obj):
        if count<3:
            L.append(i['name'])
            count=count+1
        else:
            break
    return L
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
movies['crew'][0]
```

```python
def extract_director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew']=movies['crew'].apply(extract_director)
```

```python
# Keep the top 3,000 Indian movies that actually have content in tags
indian_movies['tag_len'] = indian_movies['tags'].apply(len)
indian_movies_filtered = indian_movies.sort_values(by='tag_len', ascending=False).head(3000)
```

```python
all_movies_df = pd.concat([indian_movies_filtered, english_movies], ignore_index=True)
```

```python
import re
def normalize_title(name: str) -> str:
    # Remove special characters except letters, numbers, and spaces
    name = re.sub(r'^[^a-zA-Z0-9\s]+', '', name)
    return name
```

```python
all_movies_df['Name'].head()
```

```python
all_movies_df['Name']=all_movies_df['Name'].apply(normalize_title)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(stop_words='english',max_features=5000)
```

```python
vectors=cv.fit_transform(all_movies_df['tags']).toarray()
```

```python
vectors
```

```python
vectors[0]
```

```python
print(list(cv.get_feature_names_out()))
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
from sklearn.decomposition import TruncatedSVD

# 1. Initialize SVD (reducing 5000 features to 100-500 is common)
# n_components=100 usually captures enough variance for 20k movies
svd = TruncatedSVD(n_components=100, random_state=42)

# 2. Transform your original vectors into the reduced space
vectors_reduced = svd.fit_transform(vectors)

# 3. Now calculate similarity using the reduced vectors
similarity = cosine_similarity(vectors_reduced)
```

```python
similarity[1]
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movie_index=all_movies_df[all_movies_df['Name']==movie].index[0]
    distances=similarity[movie_index]
    movie_list=sorted(list(enumerate(distances)),reverse=True, key=lambda x:x[1])[1:9]

    for i in movie_list:
        print(all_movies_df.iloc[i[0]].Name)
```

```python
recommend('Avatar')
```
